```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nRowsRead = 1000

# NYPD_Complaint_Data_Historic.csv has more rows (> 7 million reports), but we are only loading/previewing the first 1000 rows

df = pd.read_csv('NYPD_Complaint_Data_Historic2copy.csv')
df.dataframeName = 'NYPD_Complaint_Data_Historic2copy.csv'
nRow, nCol = df.shape
```

```python
#dropping rows with NaN values
df.dropna(subset=['Y_COORD_CD','X_COORD_CD','Latitude','Longitude','CRM_ATPT_CPTD_CD','CMPLNT_FR_TM','Lat_Lon','CMPLNT_FR_DT','BORO_NM','OFNS_DESC'], inplace=True)

#dropping columns that are not significant for future data exploration
df.drop(['PARKS_NM','STATION_NAME','TRANSIT_DISTRICT','HADEVELOPT','HOUSING_PSA'],axis='columns', inplace=True)

df.drop(['PREM_TYP_DESC','SUSP_AGE_GROUP','SUSP_SEX','SUSP_RACE','JURISDICTION_CODE'], axis='columns', inplace=True)

df.drop(['ADDR_PCT_CD','PD_CD','PD_DESC','PATROL_BORO','CMPLNT_TO_DT','CMPLNT_TO_TM'], axis='columns', inplace=True)

```

```python
#replacing all NaN values in 'LOC_OF_OCCUR_DESC' with UNKNOWN
df.fillna({'LOC_OF_OCCUR_DESC':'UNKNOWN'}, inplace=True)

#replacing all NaN values in 'VIC_RACE' with UNKNOWN
df.fillna({'VIC_RACE':'UNKNOWN'}, inplace=True)

#replacing all NaN values in 'VIC_AGE_GROUP' with UNKNOWN
df.fillna({'VIC_AGE_GROUP':'UNKNOWN'}, inplace=True)

#replacing all NaN values in 'VIC_SEX' with UNKNOWN
df.fillna({'VIC_SEX':'UNKNOWN'}, inplace=True)


#check the changes
df.head(5)
```

```python
# Descriptions of the crimes are quite hard to understand, I wanted to break it down to layman's terms
print(df.OFNS_DESC.unique())
# Renaming the crimes so that they are easier to understand/ more descriptive
df_clean = df.replace({'HARRASSMENT 2': 'HARASSMENT', 'ESCAPE 3': 'ESCAPE','ASSAULT 3 & RELATED OFFENSES':
    'ASSAULT & RELATED OFFENSES','CRIMINAL MISCHIEF & RELATED OF': 'CRIMINAL MISCHIEF','OFF. AGNST PUB ORD SENSBLTY &':
    'OFFENSES AGAINST PUBLIC ORDER/ADMINISTRATION', 'OTHER STATE LAWS (NON PENAL LA': 'OTHER STATE LAWS (NON PENAL LAW)',
                       'ENDAN WELFARE INCOMP': 'ENDANGERING WELFARE OF INCOMPETENT',
'AGRICULTURE & MRKTS LAW-UNCLASSIFIED': 'AGRICULTURE & MARKETS LAW','DISRUPTION OF A RELIGIOUS SERV': 'DISRUPTION OF A RELIGIOUS SERVICE',
                       'LOITERING/GAMBLING (CARDS, DIC': 'GAMBLING', 'OFFENSES AGAINST MARRIAGE UNCL': 'OFFENSES AGAINST MARRIAGE', 'HOMICIDE-NEGLIGENT,UNCLASSIFIE': 'HOMICIDE-NEGLIGENT','E': 'UNKNOWN',
                       'D': 'BUSINESS/ORGANIZATION','F': 'FEMALE','M': 'MALE'}).copy()
```

```python
df_clean.OFNS_DESC.unique()
```

```python
df_clean.OFNS_DESC.value_counts().iloc[:10]
```

```python
# Horizontal Bar Chart of Crime Types in 10's of thousands e.g. 0.2 = 20,000 
df_clean.OFNS_DESC.value_counts().iloc[:10].sort_values().plot(kind="barh", title = "Types of Crimes",color='limegreen')
plt.show()
```

```python
# Pie Chart of the three levels of offense in America; misdemeanor, violation and felony
df_clean['LAW_CAT_CD'].value_counts().sort_values().plot(kind='pie', subplots=True, colors=['seagreen','mediumseagreen','springgreen'], shadow = True,startangle=45,
figsize=(10,5), autopct='%1.1f%%', explode=(0,0.1,0), labels=['Violation','Felony','Misdemeanor'], title = 'Level of Offense')
plt.show()
```

```python
# Graph showing the total number of crimes per year over the years
df_clean['year'] = df_clean['RPT_DT'].str.findall(r'(\d{4})').apply(lambda x: x[0])
df_clean['year'].value_counts().plot(kind="line", color=['darkolivegreen'], title = "Total Crime Reports by Year")
```

```python
#reports by month
df_clean['month'] = df_clean['RPT_DT'].str.findall('(\d{2})').apply(lambda x: x[0])
df_clean.groupby('month').size().plot(kind = 'bar', title ='Total Crime Reports by Month', color = 'olivedrab',rot=0)
```

```python
#reports by hour
df_clean['time'] = df_clean['CMPLNT_FR_TM'].str.findall('(\d{2})').apply(lambda x: x[0])
df_clean.groupby('time').size().plot(kind = 'bar', title ='Total Crime Events by Day', color = 'chartreuse', xlabel = 'hours',rot=0)

#descending order
#df['time'].value_counts().plot(kind="bar", xlabel = "hours", ylabel = "reports", rot=0)
```

```python
df_clean['BORO_NM'].value_counts().sort_values().plot(kind="barh", 
        color = 'limegreen',title = 'Total of Crime Events by Borough')
```

```python
# Analysing the Assault section -
assault_crimes = df_clean[df.OFNS_DESC.str.contains('ASSAULT')]
assault_crimes.head()
```

```python
#plot graph
assault_crimes.groupby('year').size().plot(kind = 'barh', color = 'forestgreen', title = 'Assault Reports by Year')

#average assault reports per year
mean = assault_crimes['year'].value_counts().mean()
print(round(mean,2))
```

```python
#plot graph
assault_crimes.groupby('time').size().plot(kind = 'bar', color = 'yellowgreen', rot=0, title = 'Assault Reports by Hour')
```

```python
# Graph of Assaults in Each Borough
assault_crimes['BORO_NM'].value_counts().iloc[:5].plot(kind="bar",  color = 'mediumspringgreen', title = 'Assault Crime in Each Borough',rot=45)

```

```python
#plot graph
assault_crimes['VIC_SEX'].value_counts().iloc[:3].plot(kind="bar",  color = 'darkseagreen', title = 'Assault Crime Victims by Gender', rot=0)

#victims by gender percentage
vic_sex_per = assault_crimes['VIC_SEX'].value_counts() / assault_crimes['VIC_SEX'].shape[0] *100
print(vic_sex_per)
```

```python
#plot graph
assault_crimes['VIC_AGE_GROUP'].value_counts().iloc[[0,1,2,4,5]].plot(kind="bar", color = '#2ECC71', title = 'Assault Victims by Age Group', rot=0)

#victims by age percentage
vic_age_per = assault_crimes['VIC_AGE_GROUP'].value_counts().iloc[[0,1,2,4,5]] / assault_crimes['VIC_AGE_GROUP'].shape[0] *100
print(vic_age_per)
```

```python
#plot graph
assault_crimes['VIC_RACE'].value_counts().iloc[:7].plot(kind="barh", color = 'darkkhaki', title = 'Assault Victims by Race')

#victims by race percentage
vic_race_per = assault_crimes['VIC_RACE'].value_counts().iloc[:7] / assault_crimes['VIC_RACE'].shape[0] *100
print(vic_race_per)
```
